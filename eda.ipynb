```python
# drop the missing records for products 
data.dropna(subset=["product"])
```

```python
data.isnull().sum()
```

```python
data=data.dropna(subset=["product"])
data
```

```python
data.isnull().sum()
```

```python
# replace missing item price with mean
mean_price = data["itemprice"].mean()
data["itemprice"].fillna(mean_price, inplace=True)
```
